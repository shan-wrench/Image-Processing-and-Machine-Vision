**Intensity** **Windowing** ( VS Code)

```python
# Intensity Windowing
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

c = np.array([(100, 50), (150, 200)])

t1 = np.linspace(0, c[0,1], c[0,0] + 1 - 0).astype('uint8')
print(len(t1))
t2 = np.linspace(c[0,1] + 1, c[1,1], c[1,0] - c[0,0]).astype('uint8')
print(len(t2))
t3 = np.linspace(c[1,1] + 1, 255, 255 - c[1,0]).astype('uint8')
print(len(t3))

transform = np.concatenate((t1, t2), axis=0).astype('uint8')
transform = np.concatenate((transform, t3), axis=0).astype('uint8')
print(len(transform))

fig, ax = plt.subplots()
ax.plot(transform)
ax.set_xlabel(r'Input, $f(\mathbf{x})$')
ax.set_ylabel('Output, $\mathrm{T}[f(\mathbf{x})]$')
ax.set_xlim(0,255)
ax.set_ylim(0,255)
ax.set_aspect('equal')
plt.show()


img_orig = cv.imread('/content/drive/MyDrive/Images/Emma.jpg', cv.IMREAD_GRAYSCALE)
cv.namedWindow("Image", cv.WINDOW_AUTOSIZE)
cv.imshow("Image", img_orig)
cv.waitKey(0)
image_transformed = cv.LUT(img_orig, transform)
cv.imshow("Image", image_transformed)
cv.waitKey(0)
cv.destroyAllWindows()
```

**Gamma Correction**

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

f = cv.imread('/content/drive/MyDrive/Images/Fig0309(a)(washed_out_aerial_image).tif', cv.IMREAD_GRAYSCALE)
assert f is not None

gamma = 0.5
t = np.array([(i/255)**gamma*255 for i in range(256)], dtype=np.uint8)

g = t[f] # can use g = cv.LUT(f,t)

fig,ax = plt.subplots(1,3,figsize=(15,5))
ax[0].plot(t,'r', label = "Transformation")
ax[1].imshow(f,cmap ='gray' , vmin = 0, vmax = 255)
ax[1].set_title('Original Image')
ax[1].axis('off')
ax[2].imshow(g, cmap = 'gray', vmin = 0, vmax = 255)
ax[2].set_title('Transformed with $\gamma = {0}$'.format(gamma))
ax[2].axis('off')
plt.show()
```

```python
#Histogram Equalization using opencv
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

f = cv.imread('/content/drive/MyDrive/Images/shells.tif', cv.IMREAD_GRAYSCALE)
assert f is not None

g = cv.equalizeHist(f)

fig,ax = plt.subplots(1,2,figsize=(10,5))
ax[0].imshow(f,cmap ='gray' , vmin = 0, vmax = 255)
ax[0].set_title('Original Image')
ax[0].axis('off')
ax[1].imshow(g, cmap = 'gray', vmin = 0, vmax = 255)
ax[1].set_title('Equalized Image')
ax[1].axis('off')
plt.show
```

```python
#Histogram Equalization using opencv
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

f = cv.imread('/content/drive/MyDrive/Images/shells.tif', cv.IMREAD_GRAYSCALE)
assert f is not None

h = cv.calcHist([f],[0], None,[256],[0,256])
cdf = np.cumsum(h)
L = 256
M,N = f.shape
t = np.uint8((L-1)*cdf/(M*N))
g= t[f]


fig,ax = plt.subplots(1,2,figsize=(10,5))
ax[0].imshow(f,cmap ='gray' , vmin = 0, vmax = 255)
ax[0].set_title('Original Image')
ax[0].axis('off')
ax[1].imshow(g, cmap = 'gray', vmin = 0, vmax = 255)
ax[1].set_title('Equalized Image')
ax[1].axis('off')
plt.show
```

**Average Filtering**

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

f = cv.imread('/content/drive/MyDrive/Images/sigiriya.jpg', cv.IMREAD_REDUCED_GRAYSCALE_2)
assert f is not None

k_s = 7
kernel = np.ones((k_s,k_s), dtype=np.float32)/(k_s*k_s)
g = cv.filter2D(f, cv.CV_32F, kernel)

fig,ax = plt.subplots(1,2,figsize=(10,5))
ax[0].imshow(f,cmap ='gray' , vmin = 0, vmax = 255)
ax[0].set_title('Original Image')
ax[0].axis('off')
ax[1].imshow(g, cmap = 'gray', vmin = 0, vmax = 255)
ax[1].set_title('Average Filtered Image')
ax[1].axis('off')
plt.show

```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

f = cv.imread('/content/drive/MyDrive/Images/sigiriya.jpg', cv.IMREAD_REDUCED_GRAYSCALE_2)
assert f is not None

sobel_y = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
sobel_x = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])

g_x = cv.filter2D(f, cv.CV_32F, sobel_x)
g_y = cv.filter2D(f, cv.CV_32F, sobel_y)

fig,ax = plt.subplots(1,3,figsize=(15,5))
ax[0].imshow(f,cmap ='gray' , vmin = 0, vmax = 255)
ax[0].set_title('Original Image')
ax[0].axis('off')
ax[1].imshow(g_x, cmap = 'gray', vmin = 0, vmax = 255)
ax[1].set_title('Sobel X')
ax[1].axis('off')
ax[2].imshow(g_y, cmap = 'gray', vmin = 0, vmax = 255)
ax[2].set_title('Sobel Y')
ax[2].axis('off')
plt.show

```

```python
from tkinter import Y
import numpy as np
import matplotlib.pyplot as plt

sigma = 10
hw = 3*sigma
x = np.linspace(-hw,hw,2*hw+1)
y = np.linspace(-hw,hw,2*hw+1)
X,Y = np.meshgrid(x,y)

G = 1/(2*np.pi)*np.exp(-(X**2 + Y**2)/(2*sigma**2))

#plot in 3-D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X,Y,G , cmap='coolwarm')
plt.show()
```
